```python
import plotly.graph_objects as go
import numpy as np
from geopy.distance import geodesic
from geopy import Point

points = [(np.random.uniform(0, 90), np.random.uniform(0, 180)) for _ in range(100)]

# Create the plot
fig = go.Figure()

for point in points:
    # Define the center of the circle and the radius
    latitude = point[0]
    longitude = point[1]
    radius_km = 500
    
    # Function to calculate the points of the circle
    def calculate_circle_points(lat, lon, radius_km, num_points=360):
        angles = np.linspace(0, 360, num_points)
        circle_points = []
    
        for angle in angles:
            destination = geodesic(kilometers=radius_km).destination(Point(lat, lon), angle)
            circle_points.append((destination.latitude, destination.longitude))
    
        return circle_points
    
    # Generate points for the circle
    circle_points = calculate_circle_points(latitude, longitude, radius_km)
    circle_latitudes, circle_longitudes = zip(*circle_points)
    
    # Add the filled circle
    fig.add_trace(go.Scattergeo(
        lat=circle_latitudes,
        lon=circle_longitudes,
        fill='toself',
        mode='lines',
        name='Circle',
        marker=dict(
            size=0,
        ),
        line=dict(
            width=0,
            color='rgba(0, 0, 0, 0)',
        ),
        fillcolor='blue',
        opacity=1,
    ))
    
    # Update the layout of the plot
    fig.update_layout(
        title='Geographic Circle of Radius 5 km',
        geo=dict(
            projection_type='equirectangular',
            showland=True
        )
    )
    
# Show the plot
fig.show()
```
